```python
import pandas as pd

earthquakes_df = pd.read_csv('earthquakes.csv')
earthquakes_df.head()
```

```python
earthquakes_df.describe()
```

```python
earthquakes_df.isna().sum()
```

```python
earthquakes_df = earthquakes_df.dropna()
earthquakes_df.isna().sum()
```

```python
earthquakes_df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(earthquakes_df['magnitude'], bins=40, kde=True)
plt.title('Earthquake Magnitudes')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.histplot(earthquakes_df['subnational'], bins=40, kde=True)
plt.title('Earthquake State in the USA')
plt.xlabel('The State in the USA')
plt.ylabel('Frequency')
plt.show()
```

```python
earthquakes_df['date'] = pd.to_datetime(earthquakes_df['date'])
earthquakes_timeline = earthquakes_df.set_index('date').resample('ME').size()
earthquakes_timeline
```

```python

plt.figure(figsize=(14, 7))
earthquakes_timeline.plot()
plt.title('Number of Earthquakes Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Earthquakes')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
tsunami = earthquakes_df['tsunami'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(tsunami, labels=['No Tsunami Risk Occured', 'Tsunami Risk Occured'], autopct='%1.1f%%', colors=['lightgreen', 'red'])
plt.title('Tsunami Risk')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(earthquakes_df['felt'], bins=30)
plt.title('Distribution of Felt Reports')
plt.xlabel('Number of People Who Felt the Earthquake')
plt.ylabel('Frequency')
plt.show()
```
